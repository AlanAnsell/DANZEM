```python
def HVDCFlow(sender_load, sender_gen, rec_load, rec_gen):
    if rec_gen > rec_load:
        return 0.0
    return sender_gen - sender_load

def AddFeatures(df):
    df['Total Load (MW)'] = df['NI Load (MW)'] + df['SI Load (MW)']
    df['Total Dispatch (MW)'] = df['NI Dispatch (MW)'] + df['SI Dispatch (MW)']
    df['Extra Dispatch'] = df['Total Dispatch (MW)'] - df['Total Load (MW)']
    df['Extra Dispatch %'] = 100.0 * df['Extra Dispatch'] / df['Total Load (MW)']
    df['SI-NI HVDC (MW)'] = [HVDCFlow(df['SI Load (MW)'].loc[i], df['SI Dispatch (MW)'].loc[i],
                                      df['NI Load (MW)'].loc[i], df['NI Dispatch (MW)'].loc[i])
                             for i in df.index]
    df['NI-SI HVDC (MW)'] = [HVDCFlow(df['NI Load (MW)'].loc[i], df['NI Dispatch (MW)'].loc[i],
                                      df['SI Load (MW)'].loc[i], df['SI Dispatch (MW)'].loc[i])
                             for i in df.index]
    df['Total HVDC (MW)'] = [max(df['SI-NI HVDC (MW)'].loc[i], df['NI-SI HVDC (MW)'].loc[i])
                             for i in df.index]
    df['HVDC Squared'] = df['Total HVDC (MW)'] * df['Total HVDC (MW)']
    
def Exemplify(df, multiplier, label):
    df['NI Dispatch (MW)'] *= multiplier
    df['SI Dispatch (MW)'] *= multiplier
    df['label'] = label
    df.index = ['%s@%d' % (i, label) for i in df.index]

positive = df.copy()
negative = df.copy()
Exemplify(positive, 1.005, 1.0)
AddFeatures(positive)
Exemplify(negative, 0.995, 0.0)
AddFeatures(negative)

positive.head()
```

```python
from sklearn import model_selection
from sklearn import linear_model
from sklearn import preprocessing

scaler = preprocessing.MinMaxScaler()
X = pd.DataFrame(scaler.fit_transform(X),
                 index=X.index,
                 columns=X.columns)

X.head()
```

```python
from DANZEM import model

def lr_model_fn(X):
    return linear_model.LogisticRegression()

def sklearn_eval_fn(model_, X, y):
    pred = model_.predict(X)
    return model.Result(model_, pred, y, model_.score(X, y))

def sklearn_train_fn(model_fn, eval_fn, X, y, process_fn=None, verbose=False):
    model_ = model_fn(X)
    model_.fit(X, y)
    return model_, 0.0
```
