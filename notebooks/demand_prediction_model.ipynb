```python
from DANZEM import data
from DANZEM import utils
from DANZEM import lgp

utils.SetDir('C:\\Users\\tahjaa\\Documents\\Research\\DANZEM\\DANZEM\\data')

def GetDemand(day):
    df = data.GetLGPFile(day)
    return lgp.ExtractLoadInfo(df)

demand_df = data.GetDataForRange((2013, 1, 1), (2017, 12, 31), GetDemand)
demand_df.head()
```

```python
tp_df = tps[['dst_begins', 'dst_ends', 'is_national_holiday', 'is_observed_holiday', 'yearday']]

utils.OneHot(tps, columns=['month', 'time_period'], add_df=tp_df)
tp_df.head()
```

```python
X = tp_df[5:].copy()
for col in list(demand_df):
    X[col] = list(demand_df[col][:-5])
    
X.head()
```

```python
y = demand_df['NI Total Load (MW)'][5:].copy()
y.head()
```

```python
from sklearn import preprocessing

import pandas as pd

scaler = preprocessing.MinMaxScaler()
X = pd.DataFrame(scaler.fit_transform(X),
                 index=X.index,
                 columns=X.columns)

X.head()
```

```python
from DANZEM import model

model_fn = model.MakeNN([50, 30, 20, 10, 5, 3], loss='mae', reg_strength=0.0001)

performance, results = model.CrossEvaluate(X, y, model_fn,
                                           loss='mae',
                                           model_type = model.KERAS,
                                           max_val_stagnation_epochs=6,
                                           verbose=True)
```
