```python
tps = data.TPRange((2013, 1, 1, 1), (2017, 1, 1, 1))
len(tps)
```

```python
tps = tps[::2]
tps[:5]
```

```python
def TPToTime(tp):
    year, month, day, tpnum = tp
    hour = (tpnum - 1) // 2
    return '%02d/%02d/%d %02d:00' % (day, month, year, hour)

times = [TPToTime(tp) for tp in tps]
times[:5]
```

```python
sorted(list(set(times) - set(temps['Date(NZST)'])))
```

```python
sorted(list(set(temps['Date(NZST)']) - set(times)))
```
